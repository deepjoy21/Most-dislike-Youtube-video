# Step 5 : Input data Visualisation

```python
df= df['Uploader / artist'].value_counts().plot(kind='bar')
plt.title ('Total number of videos uploaded by various Uploader/artist')
plt.xlabel('Names of Uploader/artist')
plt.ylabel('Total number of videos')
plt.figure()
```

```python
df= df['Type of video (Style)'].value_counts().plot(kind='bar')
plt.title ('Types of videos uploaded by various Uploader/artist')
plt.xlabel('Type of videos')
plt.ylabel('Total number of videos')
plt.figure()
```

```python
df= df['Dislikes (%)[B]'].value_counts().plot(kind='bar')
plt.title ('Total number of dislikes in videos uploaded')
plt.xlabel('Percentage of dislikes')
plt.ylabel('Total number of dislikes')
plt.figure()
```

# Step 6: Exploratory Data Analysis



```python
ax = sb.countplot(x = 'Uploader / artist',data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sb.countplot(x = 'Type of video (Style)',data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```
